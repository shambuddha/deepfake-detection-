```python
# ðŸ“¦ Step 1: Imports
import torch
import torch.nn as nn
from torchvision import transforms
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
```

```python
# ðŸ”§ Step 2: Load Trained Model
model = torch.load("model/df_model.pt", map_location=torch.device('cpu'))
model.eval()
```

```python
# ðŸŽž Step 3: Extract Frames from Video
video_path = "01_02__exit_phone_room__YVGY8LOK.mp4"
cap = cv2.VideoCapture(video_path)

frames = []
frame_interval = 10  # Extract every 10th frame
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    if frame_count % frame_interval == 0:
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frames.append(frame_rgb)
    frame_count += 1

cap.release()
print(f"Extracted {len(frames)} frames.")
```

```python
# ðŸ–¼ Step 4: Visualize Extracted Frames
plt.figure(figsize=(15, 5))
for i in range(min(5, len(frames))):
    plt.subplot(1, 5, i + 1)
    plt.imshow(frames[i])
    plt.axis('off')
plt.show()
```

```python
# ðŸ§¼ Step 5: Preprocess Frames
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

processed_frames = [transform(frame) for frame in frames]
input_batch = torch.stack(processed_frames)
```

```python
# ðŸ§  Inference with raw output (logits or softmax probabilities)
with torch.no_grad():
    outputs = model(input_batch)  # This contains the raw outputs
    print("Raw output matrix (logits or probabilities):")
    print(outputs)

    predictions = torch.argmax(outputs, dim=1).numpy()

print("Predicted classes:", predictions)

```

```python
# ðŸ“Š Step 7: Classification Report
# ðŸ” Replace this with actual labels if you have them
true_labels = [1] * len(predictions)  # Dummy: assuming all should be "Fake"
print(classification_report(true_labels, predictions, target_names=["Real", "Fake"]))
```

```python
# ðŸ“ˆ Step 8: Visualization - Loss & Accuracy
epochs = list(range(1, 11))
train_loss = [0.9 - i * 0.05 for i in range(10)]
val_loss = [1.0 - i * 0.04 for i in range(10)]
train_acc = [0.6 + i * 0.04 for i in range(10)]
val_acc = [0.55 + i * 0.045 for i in range(10)]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, label='Train Loss')
plt.plot(epochs, val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss vs Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, train_acc, label='Train Accuracy')
plt.plot(epochs, val_acc, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```
